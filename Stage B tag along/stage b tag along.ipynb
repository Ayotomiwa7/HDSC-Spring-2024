# Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

# Tag Along

```python
df = pd.read_csv("/content/drive/MyDrive/HDSC Spring 2024/Stage B tag along/energydata_complete.csv")
```

```python
y = np.asanyarray(df['T6'])
y = y.reshape(-1, 1)
```

```python
x = np.asanyarray(df['T2'])
x = x.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
y_pred = lin_model.predict(X_test)
y_pred
```

```python
rmse_value = np.sqrt(mean_squared_error(y_test, y_pred))
rounded_rmse = round(rmse_value, 3)
print("RMSE:", rounded_rmse)
```

```python
df.drop(columns=['date', 'lights'], inplace=True)
```

```python
x = df.drop(['Appliances'],axis=1)
y = df['Appliances']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_normalized, y_train)
```

```python
predicted_values = model.predict(X_train_normalized)
predicted_values
```

```python
mae = mean_absolute_error(y_train, predicted_values)
round(mae, 3)
```

```python
rmse = np.sqrt(mean_squared_error(y_train, predicted_values))
round(rmse, 3)
```

```python
predicted_values_test = model.predict(X_test_normalized)
mae = mean_absolute_error(y_test, predicted_values_test)
round(mae, 3)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values_test))
round(rmse, 3)
```

```python
ridge_model = Ridge()
ridge_model.fit(X_train_normalized, y_train)
```

```python
predicted_values_ridge = ridge_model.predict(X_test_normalized)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values_ridge))
round(rmse, 3)
```

```python
lasso = Lasso()
lasso.fit(X_train, y_train)
```

```python
predicted_values_lasso = lasso.predict(X_test)
predicted_values_lasso
```

```python
def get_weights_df(model, data, col_name):
    weights = pd.Series(model.coef_, data.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
lasso_weights = get_weights_df(lasso, X_train, 'Lasso_Model_weight')
lasso_weights
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values_lasso))
round(rmse, 3)
```
