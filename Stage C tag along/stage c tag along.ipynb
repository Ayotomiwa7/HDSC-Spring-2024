# Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RandomizedSearchCV
```

# Tag Along

```python
df = pd.read_csv("/content/drive/MyDrive/HDSC Spring 2024/Stage C tag along/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
df.isnull().sum()
```

```python
df[pd.isnull(df['TotalCharges'])]
```

```python
df['TotalCharges'].fillna(0, inplace=True)
```

```python
df[pd.isnull(df['TotalCharges'])]
```

```python
df.isnull().sum()
```

```python
df['Churn'].replace({'Yes':1,'No':0},inplace=True)
```

```python
df['Churn'].head()
```

```python
categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies','Contract', 'PaperlessBilling', 'PaymentMethod']
numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
df = df.drop(['customerID'], axis = 1)
df.head()
```

```python
scaler = StandardScaler()
df[numerical] = scaler.fit_transform(df[numerical])
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoded_categorical = encoder.fit_transform(df[categorical])
```

```python
encoded_feature_names = encoder.get_feature_names_out(categorical)
```

```python
encoded_feature_names
```

```python
encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=encoded_feature_names, index=df.index)
```

```python
encoded_categorical_df.head()
```

```python
processed_df = pd.concat([df[numerical], encoded_categorical_df], axis=1)
```

```python
processed_df.head()
```

```python
processed_df.shape
```

```python
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(processed_df, y, test_size=0.2, random_state=1)
```

```python
rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Random Forest Accuracy: {accuracy_rf:.4f}')
```

```python
et_model = ExtraTreesClassifier(random_state=1)
et_model.fit(X_train, y_train)
```

```python
y_pred_et = et_model.predict(X_test)
```

```python
accuracy_et = accuracy_score(y_test, y_pred_et)
print(f'Extra Trees Accuracy: {accuracy_et:.4f}')
```

```python
xgb_model = XGBClassifier(random_state=1, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
```

```python
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f'XGBoost Accuracy: {accuracy_xgb:.4f}')
```

```python
lgbm_model = LGBMClassifier(random_state=1)
lgbm_model.fit(X_train, y_train)
```

```python
y_pred_lgbm = lgbm_model.predict(X_test)
```

```python
accuracy_lgbm = accuracy_score(y_test, y_pred_lgbm)
print(f'LightGBM Accuracy: {accuracy_lgbm:.4f}')
```

```python
model_accuracies = {
    'Model': [],
    'Accuracy': []
}
```

```python
model_accuracies['Model'].append('Random Forest')
model_accuracies['Accuracy'].append(accuracy_rf)

model_accuracies['Model'].append('Extra Trees')
model_accuracies['Accuracy'].append(accuracy_et)

model_accuracies['Model'].append('XGBoost')
model_accuracies['Accuracy'].append(accuracy_xgb)

model_accuracies['Model'].append('LightGBM')
model_accuracies['Accuracy'].append(accuracy_lgbm)

accuracy_df = pd.DataFrame(model_accuracies)
```

```python
accuracy_df
```

```python
n_estimators = [50, 100, 300, 500, 1000]
min_samples_split = [2, 3, 5, 7, 9]
min_samples_leaf = [1, 2, 4, 6, 8]
max_features = ['auto', 'sqrt', 'log2', None]

hyperparameter_grid = {
    'n_estimators': n_estimators,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'max_features': max_features
}
```

```python
et_model = ExtraTreesClassifier(random_state=1)
random_search = RandomizedSearchCV(estimator=et_model,
                                   param_distributions=hyperparameter_grid,
                                   n_iter=10,
                                   scoring='accuracy',
                                   cv=5,
                                   n_jobs=-1,
                                   verbose=1,
                                   random_state=1)

random_search.fit(X_train, y_train)

best_hyperparameters = random_search.best_params_
print("Best Hyperparameters:", best_hyperparameters)
```

```python
et_model_optimized = ExtraTreesClassifier(**best_hyperparameters, random_state=1)
et_model_optimized.fit(X_train, y_train)
y_pred_optimized = et_model_optimized.predict(X_test)
accuracy_optimized = accuracy_score(y_test, y_pred_optimized)
print(f'Optimized Extra Trees Classifier Accuracy: {accuracy_optimized:.4f}')
```

```python
feature_importances = et_model_optimized.feature_importances_

feature_importance_df = pd.DataFrame({'Feature': processed_df.columns, 'Importance': feature_importances})

feature_importance_df_sorted = feature_importance_df.sort_values(by='Importance', ascending=False)
```
